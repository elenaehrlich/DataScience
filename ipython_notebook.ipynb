## Bokeh interactive plotting

```python
from bokeh.io import output_notebook, show
output_notebook()
from bokeh.models import TapTool, CustomJS, ColumnDataSource
from random import random
from bokeh.plotting import figure

x = [random() for x in range(500)]
y = [random() for y in range(500)]
color = ["navy"] * len(x)

s = ColumnDataSource(data=dict(x=x, y=y, color=color))
p = figure(plot_width=400, plot_height=400, tools="lasso_select", title="Select Here")
p.circle('x', 'y', color='color', size=8, source=s, alpha=0.4)

s2 = ColumnDataSource(data=dict(ym=[0.5, 0.5]))
p.line(x=[0,1], y='ym', color="orange", line_width=5, alpha=0.6, source=s2)

s.callback = CustomJS(args=dict(s2=s2), code="""
    var inds = cb_obj.get('selected')['1d'].indices;
    var d = cb_obj.get('data');
    var ym = 0
    
    if (inds.length == 0) { return; }
    
    for (i = 0; i < d['color'].length; i++) {
        d['color'][i] = "navy"
    }
    for (i = 0; i < inds.length; i++) {
        d['color'][inds[i]] = "firebrick"
        ym += d['y'][inds[i]]
    }
    
    ym /= inds.length
    s2.get('data')['ym'] = [ym, ym]
    
    cb_obj.trigger('change');
    s2.trigger('change');
""")

show(p)
```

## Sci-Kit Learn (sklearn)

```python
from sklearn import datasets
iris = datasets.load_iris()

iris_target_names = pandas.DataFrame(iris.target_names, columns=['species'])
iris_target = pandas.DataFrame(iris.target, columns=['target'])
iris_target['target_name'] = iris_target['target'].map(lambda x: iris_target_names.loc[x][0])
display(iris_target_names)

iris_data = pandas.DataFrame(iris.data, columns=iris.feature_names)
iris_data.join(iris_target)
```

### Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
y_pred = gnb.fit(iris.data, iris.target).predict(iris.data)
print("Number of mislabeled points out of a total %d points : %d" % (iris.data.shape[0],(iris.target != y_pred).sum()))
```

### Linear Discriminant Analysis

```python
from sklearn.lda import LDA
clf = LDA()
clf.fit(iris.data, iris.target)
y_pred2 = clf.predict(iris.data)
print("Number of mislabeled points out of a total %d points : %d" % (iris.data.shape[0],(iris.target != y_pred2).sum()))
```
